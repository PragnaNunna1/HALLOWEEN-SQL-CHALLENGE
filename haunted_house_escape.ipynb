```python
# Path to your SQLite DB
sqlite_file = r"C:\Users\pragna.nunna_jindals\Documents\DataAnalysis\Data Folder\QUIZZ\haunted_house_escape.db"

# Connect
sqlite_conn = sqlite3.connect(sqlite_file)
```

```python
# Get list of tables
tables = sqlite_conn.execute("SELECT name FROM sqlite_master WHERE type='table';").fetchall()
tables = [t[0] for t in tables]
print("Tables in SQLite:", tables)
```

```python
import sqlalchemy as sal
engine = sal.create_engine('postgresql://postgres:5599@localhost:5432/haunted_house_escape')
conn = engine.connect()
# database is created in pgAdmin
```

```python
for table in tables:
    # Read table from SQLite
    df = pd.read_sql(f'SELECT * FROM {table}', sqlite_conn)
    
    # Write to PostgreSQL (replace if table exists)
    df.to_sql(table, engine, if_exists='replace', index=False)
    
    print(f"Table '{table}' transferred successfully!")
```
